- Engine: 배기량에 따른 중고차가격의 차이를 확인해보기 위해 수치형으로 변환
- Emission_class: Euro 1에서 Euro 6으로 갈수록 더 최신에 만들어졌으며, 엄격한 규제가 적용된 자동차이다. 수치형 변수로 바꾸어, 이후 숫자가 커짐에 따라 중고차 가격의 변화가 있는지 확인해본다.



### 배기량(Engine)컬럼의 데이터를 수치형으로 변환


````python
#  Engine의 고유값 확인
cars['Engine'].unique()
````

````python
# Engine에서 문자열 L을 제거하여 수치형으로 변경
cars['Engine'] = cars['Engine'].str[:3].astype(float)
cars.info()
cars.head()
````

### 배출가스 등급(Emission_class)컬럼의 데이터를 수치형으로 변환


````python
#  Emission_class의 고유값 확인
cars['Emission_class'].unique()
````

````python
# Emission_class을 수치형으로 변환
cars['Emission_class'] = pd.to_numeric(cars['Emission_class'].str[-1])
cars.info()
cars.head(3)
````

- 배출가스등급에 결측치가 있어 int형으로 바로 변환하지 못하고 pd.to_numeric()을 이용하여 float형으로 변환했다.

- 스피어만 상관계수를 구했을 때 상관계수의 절댓값이 올라간 변수들이 있다. 이들 사이에 비선형적인 관계가 있다고 볼 수 있다.
- 가격과 상관관계가 있다고 볼 수 있는 변수는 총 주행거리, 등록년도, 배출가스 등급이다.
- 총 주행거리는 등록년도, 배출가스등급, 가격과 음의 상관관계가 있다고 볼 수 있다. 생각해보면 구입한지 시간이 지난만큼 더 많이 주행했을 것이고 오래된 자동차의 배출가스등급도 더 낮게 측정되었을테니 말이다.
- 차량의 가격, 등록년도, 배출가스등급은 서로 강한 양의 상관관계가 있음을 확인할 수 있다.

- Price: 타겟변수로 사용될 수 있으므로 가급적 변환하지 않는다.

## 주성분 분석(PCA)

````python
# 주성분분석
from sklearn.decomposition import PCA

# 10개의 주성분 추출
pca = PCA(10)
cars_pca = pca.fit_transform(cars)

cars_pca = pd.DataFrame(cars_pca, columns = list('PC' + str(x+1) for x in range(cars_pca.shape[1])))

cars_pca.head()
````

````python
# pca설명력 확인
pca.explained_variance_ratio_
````

````python
# pca누적설명력 확인
explained_ratio_table = pd.DataFrame({'var_ratio': pca.explained_variance_ratio_,
                                      'cum_var': np.cumsum(pca.explained_variance_ratio_)})

explained_ratio_table
````
